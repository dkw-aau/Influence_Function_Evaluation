```python
import tensorflow_datasets as tfds

# Load the ImageNet dataset
ds, info = tfds.load('imagenet2012', split='train', with_info=True)

# Define the categories to search for
categories = ['dog', 'fish']

# Find images labeled with both categories
matching_images = []
for example in ds:
    if all(category in example['label'].numpy().decode('utf-8') for category in categories):
        matching_images.append(example['image'].numpy())

# Print the number of matching images found
print(f'Found {len(matching_images)} matching images')
```
